1. What is mutability in Python?
  - Mutability in Python refers to the ability of an object to change its data or value after it has been created. If an object can be modified without creating a new object, it is called mutable. If an object cannot be changed once it is created, it is called immutable.
Examples of mutable objects include lists, dictionaries, and sets.

3. What are logical operators in Python?
 - Logical operators in Python are used to combine multiple conditions or expressions and return a Boolean result, which is either True or False. They are commonly used in decision-making statements and loops.

Python provides three main logical operators:

AND: Returns true if all conditions are true.

OR: Returns true if at least one condition is true.

NOT: Reverses the result of a condition.

Logical operators help in writing complex conditions in a simple and readable manner. They are widely used in control flow statements such as if, while, and for loops.

5. Describe how dictionaries store data.
- A dictionary in Python stores data in the form of key-value pairs. Each key is unique and is used to access its corresponding value. Keys must be immutable, while values can be of any data type.

Dictionaries use a hashing mechanism, which allows very fast data retrieval. Instead of accessing elements using index positions, data is accessed using meaningful keys.

They are commonly used to represent structured and real-world data such as student details, employee records, and configuration settings.

7.  What is the difference between for and while loops?
- The for loop is used when the number of iterations is known in advance. It is commonly used to iterate over sequences like lists, tuples, or ranges.

The while loop is used when the number of iterations is not known beforehand. It continues executing as long as a specified condition remains true.

In short, for loops are count-controlled loops, while while loops are condition-controlled loops.

8. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
- num = int(input("Enter a number: "))
factorial = 1

while num > 0:
    factorial = factorial * num
    num = num - 1

print("Factorial is:", factorial)

9. Why might you use a set instead of a list in Python?
- A set is used instead of a list when uniqueness of elements is required. Sets automatically remove duplicate values, while lists allow duplicates.

Sets also provide faster searching because they are based on hashing. However, sets do not maintain order and do not support indexing.

Sets are useful when performing mathematical operations such as union, intersection, and difference.
