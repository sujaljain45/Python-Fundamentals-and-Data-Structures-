# Functions

1. What is the difference between a function and a method in Python?
  - A function is a block of reusable code that performs a specific task and is defined independently. It can be called directly using its name and does not belong to any class.

A method is also a block of code, but it is associated with a class and operates on the data (attributes) of an object created from that class. Methods are called using an object of the class.

In simple words, a function is independent, whereas a method is dependent on an object. Functions are used for general-purpose tasks, while methods are used to define behaviors of objects in Object-Oriented Programming.

2. Explain the concept of function arguments and parameters in Python.
- In Python, parameters are variables listed in the function definition, while arguments are the actual values passed to the function when it is called.

Parameters act as placeholders that receive data, and arguments supply the actual data. Python supports different types of arguments such as positional arguments, keyword arguments, default arguments, and variable-length arguments.

This concept allows functions to be flexible and reusable, enabling the same function to work with different input values.

3. What is the purpose of the return statement in a Python function?
 - The return statement in a Python function is used to send a result back to the caller of the function. It ends the execution of the function and passes the specified value to the place where the function was called.

Without a return statement, a function returns None by default. The return statement allows functions to produce outputs that can be stored in variables or used in further calculations.

It is an essential part of writing meaningful and reusable functions in Python.

4. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
 - num = int(input("Enter a number: "))
factorial = 1

while num > 0:
    factorial = factorial * num
    num = num - 1

print("Factorial is:", factorial)

5. What is Object-Oriented Programming (OOP)?
- Object-Oriented Programming (OOP) is a programming paradigm that focuses on organizing code using objects and classes instead of functions and logic alone.

OOP helps in modeling real-world entities by combining data and behavior into a single unit called an object. It improves code reusability, scalability, and maintainability.

The main principles of OOP are encapsulation, abstraction, inheritance, and polymorphism. Python supports OOP, making it easier to develop complex and structured programs.

6. What is a class and an object in OOP?
- A class is a blueprint or template that defines the structure and behavior of objects. It contains attributes (data) and methods (functions).

An object is an instance of a class. It represents a real-world entity and can access the attributes and methods defined in the class.

In simple terms, a class is like a design, and an object is the actual thing created using that design.


7.  What is the difference between abstraction and encapsulation?
- Abstraction focuses on hiding complex implementation details and showing only essential features to the user. It helps reduce complexity and increases clarity.

Encapsulation focuses on bundling data and methods together within a class and restricting direct access to data to protect it from misuse.

In short, abstraction hides unnecessary details, while encapsulation protects data and controls access.

9. What is inheritance in OOP? Provide a simple example in Python
- Inheritance is an OOP concept where one class acquires the properties and methods of another class. The existing class is called the parent class, and the new class is called the child class.

Inheritance promotes code reusability and reduces duplication.

Code - class Animal:
    def sound(self):
        print("Animals make sounds")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.sound()
d.bark()

10.  What is a lambda function in Python? Provide an example.
- A lambda function is a small anonymous function defined using the lambda keyword. It can have any number of arguments but only one expression.

Lambda functions are commonly used for short operations where defining a full function is unnecessary.

 Code - square = lambda x: x * x
print(square(5))
